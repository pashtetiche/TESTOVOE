```python
import pandas as pd
from google.cloud import bigquery
import os
import numpy as np
import math
import statistics
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt
```

Инициализация клиента

```python
plt.style.use('seaborn-whitegrid')
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "json_key_cloud.json"
client = bigquery.Client()
```

Запрос для взятия всех статистик

```python
query_fulldata_analytics = """
with data as (
	select CONCAT(fullVisitorId, CAST(visitId AS STRING)) AS unique_session_id,
	totals.totalTransactionRevenue/1e6 AS total_revenue,
	totals.visits as visits,
	totals.hits as hits,
	totals.pageviews as pageviews,
	totals.timeOnSite as timeOnSite,
	totals.newVisits as newvisits,
	trafficSource.source as source,
	device.browser as browser,
	device.operatingSystem as operatingSystem,
	device.deviceCategory as deviceCategory,
	channelGrouping,
	case when totals.totalTransactionRevenue is not NULL then 1 else NULL end as revenue_true,
	case when totals.totalTransactionRevenue is NULL then 1 else NULL end as revenue_false,
    case when totals.totalTransactionRevenue is not NULL and totals.newVisits = 1
    then totals.totalTransactionRevenue/1e6 else NULL end as revenue_from_new
from `bigquery-public-data.google_analytics_sample.ga_sessions_*`
  WHERE
      _TABLE_SUFFIX BETWEEN '20170601' AND '20170631'
      AND geoNetwork.country = 'United States'
)

select * from 
(
select distinct channelGrouping,operatingSystem,
	count(unique_session_id) over w as count_parted,
    sum(newvisits) over w as sum_new_visits,
	count(unique_session_id) over w / count(unique_session_id) over () as count_parted_to_all,
	PERCENTILE_CONT(total_revenue, 0.5) OVER w as median_revenue,
	sum(total_revenue) OVER w as sum_revenue,
    sum(revenue_from_new) over w / sum(total_revenue) OVER w as percent_revenue_from_new,
	count(total_revenue) over w / count(unique_session_id) over w as revenue_to_parted_all,
	PERCENTILE_CONT(visits, 0.5) OVER w as median_visits,
	PERCENTILE_CONT(hits, 0.5) OVER w as median_hits,
	PERCENTILE_CONT(pageviews, 0.5) OVER w as median_pageviews,
	sum(pageviews) OVER w as sum_pageviews,
	PERCENTILE_CONT(timeOnSite, 0.5) OVER w as median_timeonsite,
    
	stddev_samp(total_revenue) over w / avg(total_revenue) over w as koef_var_revenue,
	stddev_samp(visits) over w / avg(visits) over w as koef_var_visits,
    stddev_samp(hits) over w / avg(hits) over w as koef_var_hits,
	stddev_samp(pageviews) over w / avg(pageviews) over w as koef_var_pageviews,
	stddev_samp(timeOnSite) over w / avg(timeOnSite) over w as koef_var_timeOnSite
    
	
from data
WINDOW w AS (partition by channelGrouping,operatingSystem)
   order by sum_revenue desc
   )
   where count_parted >= 500
"""
```

Вывод основных статистик

```python
query_job = client.query(query_fulldata_analytics)
rows = query_job.result()
for row in rows:
	row_d = dict(row)
	for i in row_d:
		print("{:<24} {}".format(i,row_d[i]))
	print()
```

Коэфициент вариации сложно интерпретировать, однако это один из показателей рассеивания данных по выборке. В случаях, когда будет подобран правильный теоретический закон распределения, среднее квадратическое отклонение (СКО) будет полезым для вероятностного анализа данных. Грубо говоря, если данные будут иметь нормальную, или логнормальную природу, можна будет использовать правило трёх сигм и предсказать, в каких пределах с 95% вероятностью будет находиться анализируемая выборка. Для рынка недвижимости эти границы вполне адекватны.

В общих чертах, анализ показывает, что на самых "жирных" платформах объём конверсии от старых юзеров большая, чем от новых и составляет 70-75 % от общего объёма конверсий. Но всё сильно зависит от канала рекламы и операционной системы юзера.

revenue_to_parted_all показывает, какой процент сессий были с конверсией. 8% таких сессий было у Mac по реферальной системе. При этом, сумма просмотров страниц этой выборки в два раза больше, чем у второго места по объёме сессий, Windows, хотя и сессий у Windows было всего на 500 меньше.

Понедельная временная динамика

```python
date_query = """
select date, sum(total_revenue) as revenue_sum, avg(total_revenue) as revenue_avg, 
sum(timeOnSite) as sum_timeOnSite
from (
	select CONCAT(fullVisitorId, CAST(visitId AS STRING)) AS unique_session_id,
	totals.totalTransactionRevenue/1e6 AS total_revenue,
    totals.timeOnSite as timeOnSite,
	TIMESTAMP_TRUNC(TIMESTAMP(PARSE_DATE("%Y%m%d", date)), WEEK, 'UTC') as date
from `bigquery-public-data.google_analytics_sample.ga_sessions_*`
  WHERE
      _TABLE_SUFFIX BETWEEN '20160601' AND '20170830'
      AND geoNetwork.country = 'United States'
      ) as datetime
      group by date
      order by date
"""

#перекомпоновка данных в удобный объект
query_job = client.query(date_query)
rows = query_job.result().to_dataframe()
all_data_object = {k:v for k,v in zip(rows.columns,[np.array(list(filter(lambda v: v==v, rows[i]))) for i in rows])}

plt.rc('font', size=10)
fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()
ax1.plot(all_data_object['date'], all_data_object['revenue_sum'], color='tab:orange', label='Total revenues')
#ax.plot(all_data_object['date'], all_data_object['sum_timeOnSite'], color='tab:blue', label='Total time on site')
ax2.plot(all_data_object['date'], all_data_object['revenue_avg'], color='tab:red', label='Average revenues')
ax2.set_ylabel('Median revenues')
ax2.set_xlabel('Time')
#ax2.grid(True)
ax2.legend(loc='upper right')
ax1.set_xlabel('Time')
ax1.set_ylabel('Sum revenues')
ax1.set_title('Time series')
ax1.grid(True)
ax1.legend(loc='upper left')
#ax1.xticks(all_data_object['date'], rotation='vertical')
#ax2.xticks(all_data_object['date'], rotation='vertical')
plt.margins(0.2)
plt.subplots_adjust(bottom=0.15)
plt.show()
```

По динамике можна судить, что объём конверсий растёт с медианной конверсией. Есть гипотеза, что это изза распродаж на какой-то из типов товаров.
